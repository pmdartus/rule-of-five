# Rule of five

**Definition:** _There is a 93.75% chance that the median of a population is between the smallest and largest values in any random sample of five from that population._

From Douglas Hubbard - "How to Measure Anything"

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
SAMPLE_SIZE = 5

def rule_of_five(population: list[float], median: float) -> bool:
    sample = np.random.choice(population, SAMPLE_SIZE)
    return np.min(sample) < median < np.max(sample)
```

```python
POPULATION_SIZE = 10_000_000
ITERATION = 10_000

def test_rule_of_five(population: list[float], iteration: int) -> list[bool]:
    median = np.median(population)
    return [rule_of_five(population, median) for _ in range(iteration)]
```

## Uniform distribution

```python
uniform_distribution = np.random.rand(POPULATION_SIZE)
```

```python
plt.hist(uniform_distribution, bins=100, density=True, histtype='step')

plt.title('Uniform distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.show()
```

```python
uniform_test = test_rule_of_five(uniform_distribution, ITERATION)
```

```python
cumulative_probability = np.cumsum(uniform_test) / np.arange(1, ITERATION + 1)

plt.plot(cumulative_probability)
plt.axhline(0.9375, color='r', linestyle='--')

plt.title('Rule of five test on uniform distribution')
plt.xlabel('Iteration')
plt.ylabel('Cumulative probability')

plt.show()
```
